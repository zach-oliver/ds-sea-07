### 3 Month Average Rent
One of the first and simplest ways to evaluate the rent for a specific zip code, number of bedroom, and number of bathrooms is a simple 3 month average. The query to collect that data from MySQL is below:

```python
# DO NOT RUN THIS CODE AS THIS IS MYSQL AND NOT PYTHON
SELECT 
    round(r.bedrooms) AS 'beds',# Round Bedrooms so they are whole numbers
    ceiling(r.bathrooms) AS 'baths', # Round all bathrooms up so 1 1/2 baths can be grouped with those with 2
    r.postal_code AS 'zip',
    r.locality AS 'city',
    COUNT(*) AS 'count', # Get the count of listings which are being grouped
    MIN(r.date) AS 'min date', # what's the lowest date for the listing grouped?
    MAX(r.date) AS 'max date', # what's the most recent date for the listings grouped?
    round(AVG(r.price),2) AS 'price', # round and average the price to 2 decimal points
    round(AVG(r.footage)) AS 'sq ft', # round and average the footage as a whole number
    round(AVG(r.price_sqft),2) AS 'price / sq ft' # round and average the price/sq ft
FROM
    # this section is a query within a query where the top query pulls from this sub-query
    (SELECT 
        * # get all fields
    FROM
        rentals_all_filtered
    WHERE
        fl_super_clean = TRUE # super clean means that all fields have validate input
            AND date >= (CURDATE() - INTERVAL 3 MONTH) # the date of the price is within the last 3 months
    ORDER BY date) r
GROUP BY bedrooms , bathrooms , postal_code , locality # group results of top query by bed, bath, zip, city
HAVING COUNT(*) > 10; # the groups should have at least 10 listings in a 3 month period
```

### 3 Month Linear Regression to Calculate Rents
Now that we know the average, let's see if we can get something better than the average to calculate rents as there are many different factors outside of just bed, bath, and price that determine the price.

Below is the MySQL used to collect all of the rentals from the last 3 months where they have a zip code that has a count greater than 10 in the last 3 months so we know there is enough data to calculate the linear regression.

```python
# DO NOT RUN THIS CODE AS THIS IS MYSQL AND NOT PYTHON
SELECT 
    *
FROM
    rentals_all_filtered
WHERE
    date >= (CURDATE() - INTERVAL 3 MONTH) # Get all data where price date is less than 3 months old
        AND fl_outlier = 'not_outlier' # And the price is not an outlier (aka standard deviation) to omit the times when houses for sale got crossed with rental data
        AND footage IS NOT NULL # Get data where the sq footage isn't blank
        AND lat IS NOT NULL # Get data where latitude isn't blank. You'll notice this is helpful for the linear regression
        AND lon IS NOT NULL # Get data where longitude isn't blank. You'll notice this is helpful for the linear regression
        AND postal_code IN (SELECT DISTINCT # Postal code has a postal code in this sub-query
                s.zip # Get only the distinct zip codes from this sub-query
                FROM
                    (SELECT # sub-sub-query groups beds, baths together by zip and only returns beds, baths, zip where count is > 10
                        ROUND(r.bedrooms) AS 'beds',
                            CEILING(r.bathrooms) AS 'baths',
                            r.postal_code AS 'zip',
                            COUNT(*) AS 'count'
                    FROM
                        (SELECT  # sub-sub-sub-query gets all data where price data is less than 3 months old and not a price outlier
                        *
                        FROM
                            rentals_all_filtered
                        WHERE
                            date >= (CURDATE() - INTERVAL 3 MONTH)
                                AND fl_outlier = 'not_outlier'
                        ORDER BY date) r
                    GROUP BY r.bedrooms , r.bathrooms , r.postal_code
                    HAVING COUNT(*) > 10
                    ORDER BY r.bedrooms , r.bathrooms , r.postal_code) s
                ORDER BY s.zip)
        ORDER BY postal_code , date;
```
